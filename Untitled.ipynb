```python
class Game:
    def __init__(self, num1, num2, num3):
        
        self.heaps = [num1, num2, num3]
        
        # First player starts
        # 1 denotes that the human starts first
        self.player_turn = 1
        
    def write_heaps(self):
        print('heap 1: {}'.format(self.heaps[0]))
        print('heap 2: {}'.format(self.heaps[1]))
        print('heap 3: {}'.format(self.heaps[2]))
        
    def valid(self, heap, num):
        if heap < 0 or heap > 2 or self.heaps[heap] < num:
            return False
        else:
            return True
     
    # Checking whether the game came to an end
    def end(self):
        lens = len(self.heaps)
        zero_counts = 0
        for i in range(lens):
            if self.heaps[i] == 0:
                zero_counts += 1
        if zero_counts == lens -1:
            return True
        else:
            return False
    
    # The Max player is a computer
    def Max(self):
        maxv = -2
        heap = None
        num = None

        result = self.end()
        if result:
            return (-1, None, None)
        else:
            for i in range(len(self.heaps)):
                for j in range(1, self.heaps[i]+1):
                    if self.heaps[i] > 0:
                        self.heaps[i] -= j
                        m = self.Min()
                        if m > maxv:
                            maxv = m
                            heap = i + 1
                            num = j
                        self.heaps[i] += j

        return (maxv, heap, num)
    
    
    def Min(self):
        minv = 2
        result = self.end()
        if result:
            return 1
        else:
            for i in range(len(self.heaps)):
                for j in range(1, self.heaps[i]+1):
                    if self.heaps[i] > 0:
                        self.heaps[i] -= j
                        m, _, _ = self.Max()
                        if m < minv:
                            minv = m
                        self.heaps[i] += j

        return minv
    
        
    def play(self):
        
        while True:
            self.write_heaps()

            if self.end():
                if self.player_turn == 1:
                    print('The computer won!')
                    return
                else:
                    print('The human won!')
                    return

            # if a player X (i.e., human) is on the move
            if self.player_turn == 1:

                while True:
                    heap = int(input('Enter the heap number: '))
                    num = int(input('Enter the number of balls that can be removed from the heap: '))
                    print('Human removes {} ball from heap {}'.format(num, heap))
                    heap -= 1
                    if self.valid(heap, num):
                        self.heaps[heap] -= num
                        self.player_turn = 0
                        break
                    else:
                        print("Input values are wrong, please re-enter")
                
            # if a player O (i.e., computer) is on the move
            else:
                m, heap, num = self.Max()
                print('Machine removes {} ball from heap {}'.format(num, heap))
                self.heaps[heap-1] -= num
                self.player_turn = 1
                
```

```python
g = Game(3, 4, 5)
g.play()
```
